## Train Test Split

### Now, we have to split the dataset into train and test, splitting it in such way that where testing dataset will have 20% of 11951 observations, rest will be used for training the dataset

```python
# Importing train test split from sk learn library
from sklearn.model_selection import train_test_split
# importing Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

## Model 1
- Predicting GT Compressor decay state coefficient based on 14 features.

```python
#Defining features and labels
X = mt_df.drop(['Compressor DSC','Turbine DSC'], axis = 1)
y = mt_df['Compressor DSC']
```

```python
#splitting the dataset into the train test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 95)
```

```python
print('X_train Shape',X_train.shape)
print('y_test Shape',y_train.shape)
print('X_train Shape',X_test.shape)
print('y_test Shape',y_test.shape)
```

```python
X_train.columns
```

```python
# Creating an instance of a LinearRegression()
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
pred = lr.predict(X_test)
```

```python
# Printing out the coefficients of the model
print('coefficients: ',lr.coef_)
```

```python
#Aligning the values with the parameters
coeff = pd.DataFrame(lr.coef_, X.columns)
coeff.columns = ['Coefficients']
coeff
```

```python
# Checking the RMSE and r2_score value between y_test and predicted y_test values
print('RMSE- ', np.sqrt(metrics.mean_squared_error(y_test, pred)))
print('R2 Score- ', metrics.r2_score(y_test, pred))
```

```python
# Creatinng a scatterplot of the y_test values vs the predicted y_test values
sns.set_style('darkgrid')
plt.scatter(y_test, pred, c = 'blue')
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
#importing statsmodel for multiple regression
import statsmodels.api as sm
```

```python
def Linear_Regression(x, y):
    x = sm.add_constant(x)
    #defining model, fittinf it and printing summary
    sm_model = sm.OLS(y,x).fit()
    print(sm_model.summary())
```

```python
Linear_Regression(X,y)
```

## Inference
- The model explains a significant portion of the variability in Compressor DSC (R-squared = 0.843).
- Most predictors are statistically significant, with several having strong positive or negative relationships with the dependent variable.
- Both GTT and HP do not significantly contribute to explaining the variability in Compressor DSC in the presence of other variables in the model. So, we drop those variable and check again if there is any improvement.
- The regression equation for the GT Compressor decay state coefficient (Compressor DSC) can be written as follows:   Compressor DSC = 1.5532+0.2612×Lever position−0.0766×Ship speed (v)−4.302e-07×GTT−2.952e-05×GTn+3.107e-05×GGn−0.0006×Ts−0.0006×Tp−4.174e-08×HP−0.0034×T2+0.1888×P48+0.0139×P2+0.7339×Pexh−0.0007×TIC+0.2206×mf

```python
X.columns
```

### Inference after making the iterations
- The R-squared value remains same as 84.3%.
- The p-values of all the considered features do play an important role in predicting the label.
- We can improve the model further by considering interaction terms based on the correlation heatmap.

```python
# Get the interaction terms by multiplying values
a = X['Ship speed (v)']*X['Lever position']
b = X['Tp']*X['Ts']
c = X['P48']*X['GTn']
d = X['P48']*X['Ts']
e = X['P48']*X['Tp']
f = X['P2']*X['Ts']
g = X['P2']*X['Tp']
h = X['P2']*X['GTn']
i = X['P2']*X['P48']
j = X['Pexh']*X['Tp']
k = X['Pexh']*X['Ts']
l = X['Pexh']*X['P48']
m = X['Pexh']*X['P2']
```

```python
#Adding interactions terms in X
interaction = pd.concat([X,a,b,c,d,e,f,g,h,i,j,k,l,m,],axis=1)
```

```python
interaction.head()
```

```python
#Renaming the columns
interaction = interaction.rename(columns = {0:'a',1:'b',2:'c',3:'d',4:'e',5:'f',6:'g',7:'h',
                                            8:'i',9:'j',10:'k',11:'l',12:'m'})
```

```python
interaction.head()
```

```python
Linear_Regression(interaction, y)
```

### Summary of this model
- R-squared: 0.934 and Adjusted R-squared: 0.934 indicate that the model explains 93.4% of the variance in the dependent variable, Compressor DSC. This suggests a very good fit.
- We will stop here and declare this model as the most suitable for predicting GT Compressor decay state coefficient.
